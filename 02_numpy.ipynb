```python
# numpy array의 데이터 타입 지정
import numpy as np
arr = np.array([1, 2, 3, 4, 5, 3.5])
arr.dtype

# numpy array를 생성할 때 data type을 지정할 수 있어요!!
arr = np.array([1, 2.6, 3.6, 4.9, 5.1], dtype=np.float64)
arr
# 이미 만들어진 numpy array의 데이터 타입을 변경할 수 있다!
arr = arr.astype(np.int32) # 소수점 이하를 다잘라버림
arr
```

```python
# numpy array의 다양한 생성함수
arr = np.array([1, 2, 3, 4, 5]) # basic
arr = np.zeros((3, 4)) # 3행 4열을 만드는데 0을 다 채워서 만들고 기본 데이터 타입은 float64
arr = np.ones((4, 4)) # 4행 4열을 만드는데 1을 다 채워서 만들어라
arr = np.empty((3, 3), dtype=np.int32) # 3행 3열의 공간을 확보하라 값은 보장못함
arr
```

```python
# ndarray의 자주 사용되는 생성함수
# np.arrange(시작, 끝, 간격) 시작 : inclusive, 끝 : exclusive
import numpy as np
arr = np.arange(0, 10, 1)
arr
arr = np.arange(10)
```

```python
# ndarray의 생성함수 중 하나
# np.linspace(시작, 끝, 샘플 개수)
# 주어진 범위를 지정한 샘플의 개수로 나누어 정해지는
# 간격으로 연속적인 배열 생성
# 간격 : (끝-시작) / (샘플개수 - 1)
arr = np.linspace(0, 10, 51)
arr
```

```python
# numpyarray의 shape 조절
arr = np.arange(12)
arr
arr.shape = 2, 6 # shape 을 조절할 수 있지만 이렇게 사용하는 경우는 그닥 없다.
arr

arr = np.arange(12)
arr1 = arr.reshape(4, 3) # reshape는 리턴값이 있다. 뷰 로서 원본과 연계됨
arr1
arr[0] = 100
arr1

if arr1.base is arr: # 이식의 데이터는 어디있는거야?
    print("데이터가 같아요!")
    
arr = np.arange(12)
arr1 = arr.reshape(4, 3).copy() # view를 생성하지 않고 새로운 배열을 생성

if arr1.base is arr: # 이식의 데이터는 어디있는거야?
    print("데이터가 같아요!")
else:
    print("다르다")
```

```python
# Movielens Dataset
# Group lens 라는 미네소타 대학의 컴퓨터학과 연구실에서 수집한 영화 데이터
# 특정 영화에 대한 선호를 1~5사이의 실수로 표시, 높은점수일수록 선호도가 높다

import numpy as np
data = np.loadtxt("data/movie/ratings.csv", delimiter=",", dtype=np.float64)

data.shape
data[:5,:]
# userid, movieid, rating, timestamp
```

```python
# 모든 영화의 전체 평균 평점
data[:, 2].mean() # 평균구하기
data[:, 2].sum() # 다더하기
user_ids = np.unique(data[:,0]) # 중복제외한 값 다넣기 np.unique
user_ids.shape[0] # 유저수 구하기

result = [] # python의 list
 
    
data[:,0] == 1.0 # 1과 같으면 true

result = []
for user_id in user_ids:
    data_for_user = data[data[:,0] == user_id, :]
    mean_rating_for_user = data_for_user[:,2].mean()
    result.append([user_id, mean_rating_for_user])
result
        
        
```
